2. Menghapus beberapa kolom yang tidak akan digunakan

Berdasarkan pertanyaan bisnis yang sudah ada, maka untuk kolom yang akan didrop, yakni:

- dteday : Karena indeks ini tidak memberikan informasi tambahan yang diperlukan.
- instant : Karena indeks catatan ini tidak memberikan informasi tambahan yang berguna.

```python
drop_col = ['dteday', 'instant', 'yr', 'mnth', 'weathersit', 'weekday', 'temp', 'atemp']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

3. Mengubah beberapa kolom

```python
day_df.rename(columns={
    'hum': 'humidity',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
```

```python
day_df['season'] = day_df.season.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 Bagaimana pengaruh hari libur terhadap tingkat penggunaan sepeda?

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15,10))

sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0],
    palette='pastel')
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

sns.barplot(
    x='holiday',
    y='count',
    data=day_df,
    ax=axes[1],
    palette='Set2')
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi data diatas menggunakan barplot, didapat pernyataan sebagai berikut:

- Jumlah penyewa sepeda lebih banyak pada hari kerja daripada holiday. 0 menandakan holiday dan 1 menandakan hari kerja.
- Jumlah penyewa sepeda jauh lebih banyak pada hari biasa (bukan hari libur) dibandingkan hari libur.

### Pertanyaan 2 Bagaimana pengaruh kelembaban atau kecepatan angin terhadap pola peminjaman sepeda?

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(
    x='windspeed',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Windspeed vs Count')

plt.subplot(1, 3, 2)
sns.scatterplot(
    x='humidity',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
```

Berdasarkan hasil analisis diatas menggunakan scatter plot, didapatkan pernyataan sebagai berikut:

- Hubungan antara windspeed dan count ternyata menunjukan korelasi yang negatif, karena kemiringan atau arah kecenderungan secara visual menunjukan bahwa arahnya cenderung ke kiri
- Hubungan antara hum dan count ternyata menunjukan korelasi yang sedikit negatif, tetapi tidak terlalu signifikan karena condong ke kirinya tidak terlalu nampak.

### Pertanyaan 3 Bagaimana pengaruh musim (season) terhadap penyewa sepeda berdasarkan pelanggan Casual dan Registered?

```python
seasonal_usage = day_df.groupby('season')[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['registered'],
    label='Registered',
    color='lightblue'
)

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['casual'],
    label='Casual',
    color='lightcoral'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah penyewaan sepeda berdasarkan musim')
plt.legend()
plt.show()
```

Berdasarkan hasil analisis diatas, dapat disimpulkan bahwa di semua musim (season) yang paling diminati bagi para pengguna sepeda (Casual dan Registered) adalah musim gugur (Fall), lalu diikuti dengan musim panas (Summer), musim dingin (Winter), dan diakhiri dengan musim semi (Spring).
